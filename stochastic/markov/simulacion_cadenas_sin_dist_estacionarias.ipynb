haremos el ejercicio sobre estados [[{1, 2, ..., c+d]]

```python
import matplotlib.pyplot as plt
import random
import numpy as np


def probarCadena(S, X_0, n, transiciones):
    cadena = [X_0]
    frecuencias = np.zeros(len(S))
    

    
    for k in range(n):
        #obtenemos el ulitmo elemento de la cadena como nuestro estado anterior (renglon en matriz)
        estado = random.choices(S, weights = transiciones[cadena[-1]], k = 1)[0]

        frecuencias[estado] = frecuencias[estado] + 1
        cadena.append(estado)


    
    I = [j for j in range(n+1)]

    fig, ax1 = plt.subplots()
    fig2, ax2 = plt.subplots()
    ax1.plot(I, cadena)

    #grafica de frencuencias relativas
    frecuencia_relativa = frecuencias/n
    ax2.bar(S, frecuencia_relativa)
    
```

```python
c = 3
d = 4
#nuestro conjunto de estados
S = np.arange(c + d)
print(S)

p = []
for i in S: 
    p_i = [
        1/d if i <= c and j > c else 
        1/c if i > c and j <= c else 
        0
        for j in range(c + d)
    ]
    p.append(p_i)

    
X_0 = random.randint(0, c + d)

n = 100

print(p)

probarCadena(S, X_0, n, p)


```

### Cadena de Ehrenfest

```python
d = 1000
S = np.arange(0,d+1,1) #pues podemos tener cero bolas negras en el primer contenedor


# x0 = np.random.randint(1,d) #las bolas negras iniciales
p = []

#vamos a crear los renglones de nuestra matriz de transicion
for i in range(d+1):
  p_i = [
      (i*i)/(d*d) if k == i-1 else
      (2*i*(d-i))/(d*d) if k == i else
      ((d-i)*(d-i))/(d*d) if k == i + 1 else
      0
      for k in S]
  p.append(p_i)



transiciones = np.array(p)

print(transiciones.shape)
print(len(S))

n = 1000000

for i in range(4):
    X_0 = np.random.randint(1,d) #las bolas negras iniciales
    probarCadena(S, X_0, n, transiciones)


```
