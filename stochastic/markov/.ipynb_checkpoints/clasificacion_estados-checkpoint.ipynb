### Conexión de cadenas con teoría de gráficas

*Def.* Componentes fuertemente conexos
Son las subgráficas tales que para cualesquiera dos vertices $$x,y \in G$$ existe un camino de x a y

En términos de cadenas de markov que para cualesquiera dos estados $$x,y \in S: \rho_{xy} > 0$$

Así se implementará el algoritmo de Kosaraju para encontrar todos los componentes fuertemente conexos.

Este algoritmo en dos pasadas:

1. Primera pasada: se inicia desde un vértice aleatorio un recorrido DFS y junto con una pila donde el tope representa el último vértice que llegó a su caso base, i.e. el vértice que tiene todos sus hijos visitados
2. Segunda pasada: Primero se reversan todas las flechas de la digráfica, después sobre esta se va vaciando la pila, y en cada pop de la pila se crea una componente conexa nueva si el vértice obtenido no estaba en la lista de visitados, después este vértice se agrega a los visitados y se sigue con el siguiente pop.

```python
#suponemos que p es una matriz estocastica
import random
p = np.array([])
def kosaraju(adj):
    componentes = []
    
    #parte DFS
    stack = []
    visitado = []

    for vertice in range(adj.shape[0]):
        if vertice not in visitado:
            dfs(vertice, adj, stack, visitado)



    #parte de segunda pasada
    transpuesto = adj.T #tomamos la transpuesta

    visitado = []
    while stack:
        vertice = stack.pop()
        #puede que ya lo hayamos agregado a una componente en una iteracion anterior
        if vertice in visitado:
            continue

        componente = []
        crear_componente(vertice, transpuesto, visitado, componente)
        componentes.append(componente)

    return componentes
    

#lo que buscamos es ordenar los vertices de acuerdo al recorrido dfs
def dfs(actual, adj, stack, visitado):
    visitado.append(actual)
    
    vecinos = [i for i, k in enumerate(adj[actual]) if k > 0]
    for vecino in vecinos:
        if vecino in visitado:
            continue
        dfs(vecino, adj, stack, visitado)
    #todos sus vecinos llegaron al caso base
    stack.append(actual)



def crear_componente(actual, transpuesto, visitado, componente): 
    #de una vez lo ponemos en el componente
    componente.append(actual)
    visitado.append(actual)
    
    vecinos = [i for i, k in enumerate(transpuesto[actual]) if k > 0]
    for vecino in vecinos:
        if vecino in visitado: 
            continue
        crear_componente(vecino, transpuesto, visitado, componente)
    

        
    
```

### Segunda parte del problema
Ya que tenemos las componentes fuertemente conexas, lo que queremos hacer es encontrar ciclos dirigidos en cada una de ellas. 

Para esto usamos el algoritmo de **Johnson**, suponiendo que tomamos solo una componente conexa este algoritmo consiste en: 

1. Definir tres estructuras de datos:
 - El stack donde se almacena el path del ciclo
 - Un arreglo de vertices bloqueados para saber por cuales se paso
 - Un arreglo que mapea dependencias de bloqueos, es decir  es un conjunto que responde a la pregunta:

>"Si v se desbloquea, ¿a quién más hay que desbloquear también?"

Esta última puede sonar confusa, pero pensemos en el siguiente ejemplo: 
>Queremos conseguir ciclos que inicien en w (y por definición terminen en w), así que en nuestro algoritmo llegamos a un vértice u cuyos vecinos son w y otros vértices tanto bloqueados como no bloqueados, y supongamos que nos vamos por otro vecino que no sea w, por lo que en el proceso bloqueamos al propio vértice u. Entonces así seguimos nuestro algoritmo hasta otro vértice v en el que todos los vecinos están bloqueados, entonces lo que se hace es agregar a todos esos como depenencias de v es decir si a1, ..., an son estos vertices agregamos los mapeos a1 -> v, ..., an -> v, en uno de estos mapeos esta u -> v lo que quiere decir que si en algun momento se desbloquea el vértice u entonces el v también y se abre un nuevo camino que pasa por v y posiblemente hacia u para finalizar con w.




```python
def johnson(adj)
def johnson_rec(inicio, actual, adj, stack, bloqueados, map_bloqueados, ciclos) -> bool
def unblock(actual, bloqueados, map_bloqueados)
def generar_subcadena(vertice_a_eliminar, adj)


def johnson(adj):
    #primero obtenemos las componentes conexas de kosaraju 
    componentes = kosaraju(adj)

    #despues vamos iterando por cada componente para aplicar el algoritmo de johnson
    
    for componente in componentes: 
        cantidad_vertices = adj.shape[0]
        stack = []
        bloqueados = []
        map_bloqueados = [[] for _ in range(cantidad_vertices)]
        ciclos = []

        #vamos construyendo todos los ciclos posibles que empiecen y terminen en vertice
        for vertice in componente: 
            stack.clear()
            bloqueados.clear()
            for lst in map_bloqueados:
                lst.clear()
            johnson_rec(vertice, vertice, adj, stack, bloqueados, map_bloqueados, ciclos)
            generar_subcadena(vertice, adj)
    


def johnson_rec(inicio, actual, adj, stack, bloqueados, map_bloqueados, ciclos)
    stack.append(actual)
    bloqueados.append(actual)
    vecinos = [i for i,k in enumerate(adj[actual]) if k > 0]

    ciclo_encontrado = False
    for vecino in vecinos: 
        if vecino == inicio:
            stack.append(inicio)
            ciclo = []
            ciclo.extend(reversed(stack))
            #agregar al total de ciclos
            ciclos.append(ciclo)
            stack.pop()
            ciclo_encontrado = True
            
        elif vecino not in blqueados:
            bool obtuvo_ciclo = johnson_rec(inicio, vecino, adj, stack, bloqueados, map_bloqueados, ciclos)
            
            ciclo_encontrado = ciclo_encontrado or obtuvo_ciclo

    #el momento donde se abren nuevos posibles pues sabemos que actual encontro un ciclo o que alguno de sus vecinos bloqueados 
    #lo hizo
    if(ciclo_encontrado):
        unblock(actual, bloqueados, map_bloqueados)
    else:
        #si no encontramos nigun ciclo entonces agregamos sus vecinos bloqueados al map_bloqueados
        for vecino in vecinos: 
            map_bloqueados[vecino].append(actual)

    stack.pop()
    return ciclo_encontrado
        
def unblock(actual, bloqueados, map_bloqueados)
    blqueados.remove(actual)
    if map_bloqueados[actual]: 
        for dependiente_de_actual in map_bloqueados[actual]:
            if dependiente_de_actual in bloqueados:
                unblock(dependiente_de_actual, bloqueados, map_bloqueados)
        #despues de desbloquear a todos en cadena, entonces quitamos todos los mappings que tenia dependiente_actual
        map_bloqueados[actual].clear()

#simplemente quitamos el vertice/estado de la grafica/cadena de markov, junto a los vertices salientes o entrantes
def generar_subcadena(vertice_a_eliminar, adj):
    zero = np.zeros(adj.shape[0])
    adj[vertice_a_eliminar] = zero  #le ponemos ceros para indicar que no hay flechas salientes
    adj[:,vertice_a_eliminar] = zero
```
